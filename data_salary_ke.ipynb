### Percentage of Roles That Use Python

```python
role_group["Tech_Stack"].apply(lambda x: x.str.contains("Python").sum())
```

```python
role_respondents = df["Role"].value_counts()
role_respondents
```

```python
role_python = role_group["Tech_Stack"].apply(lambda x: x.str.contains("Python").sum())
role_python
```

```python
python_users = pd.concat([role_respondents, role_python], axis="columns", sort=False)
python_users
```

```python
python_users.rename(columns={"count":"Total_Users", "Tech_Stack": "Use_Python"}, inplace=True)
python_users
```

```python
python_users["Percentage"] = (python_users["Use_Python"]/python_users["Total_Users"]) * 100
python_users
```

```python
python_users.sort_values(by="Percentage", ascending=False, inplace=True)
python_users
```

### Some Visualizations

```python
sns.countplot(x="Role", data=df, color="lightblue", order=df["Role"].value_counts().index)
plt.ylabel("Number")
plt.title("Number of Roles")
```

```python
sns.boxplot(x="Role", y="Gross_Salary", color="lightblue", showmeans=True, meanprops={'marker':'o',
                       'markerfacecolor':'white', 
                       'markeredgecolor':'black',
                       'markersize':'5'}, data=df)
plt.ylabel("Gross Salary (Ksh)")
plt.title("Salary Distribution For Different Data Roles")
```

```python
sns.boxplot(x="Gender", y="Gross_Salary", color="lightblue",width=0.5, showmeans=True, meanprops={'marker':'o',
                       'markerfacecolor':'white', 
                       'markeredgecolor':'black',
                       'markersize':'5'}, data=df)
plt.ylabel("Gross Salary (Ksh)")
plt.title("Salary Distribution Against Gender")
```

```python
order_list = ["Beginner", "1yr-2yrs", "3yrs-5yrs", "Over 5 Yrs", "Over  7 Yrs"]
#order_list=["Over  7 Yrs", "Over 5 Yrs", "3yrs-5yrs", "1yr-2yrs", "Beginner" ]
sns.barplot(x="Experience", y="Gross_Salary",color="lightblue", data=df, estimator="median",
             errorbar=None, order=order_list)
plt.ylabel("Median Salary (Ksh)")
plt.title("Median Salary According To Years of Experience")
```
